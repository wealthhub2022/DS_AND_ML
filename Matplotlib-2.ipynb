# Legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(n,m)
ax.plot(m,n)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(n,n**2)
ax.plot(n,n*3)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(n,n**2)
ax.plot(n,n*3)
ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(n,n**2, label='X**2')
ax.plot(n,n*3, label='X**3')
ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(n,n**2, label='X**2')
ax.plot(n,n*3, label='X**3')
ax.legend(loc=10)
```
