```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df= pd.read_csv('annual_gold.csv')
df.head()
```

```python
df.tail()
```

```python
# creating line chart
```

```python
plt.figure(figsize =(16,8))
sns.lineplot(data=df, x='Date', y='Gold')
```

```python
plt.figure(figsize =(16,8))
sns.lineplot(data=df, x='Date', y='Gold')
plt.xticks(rotation=30)
```

```python
plt.figure(figsize =(16,8))
sns.lineplot(data=df, x='Date', y='Gold', marker='D',markersize=15, linestyle=':')
plt.xticks(rotation=30)
```

```python
plt.figure(figsize =(16,8))
plt.plot_date(x=df['Date'], y=df['Gold'])
plt.xticks(rotation=30)
```

```python
plt.figure(figsize =(16,8))
plt.plot_date(x=df['Date'], y=df['Gold'], linestyle=':', 
              marker='X', markersize='10', c='orange') 
plt.xticks(rotation=30)
```

```python
plt.figure(figsize =(16,8))
plt.plot_date(x=df['Date'], y=df['Gold'], linestyle=':', 
              marker='X', markersize='10', c='orange',mfc='lightblue', mec='black') 
plt.xticks(rotation=30)
```

```python
plt.figure(figsize =(16,8))
plt.plot_date(x=df['Date'], y=df['Gold'], linestyle=':', 
              marker='X', markersize='10', c='orange',mfc='lightblue', mec='black') 
plt.xticks(rotation=30)
plt.title('fds')
```

```python
plt.figure(figsize =(16,8))
#Gold
plt.plot_date(x=df['Date'], y=df['Gold'], linestyle=':', 
              marker='X', markersize='10', c='orange',mfc='lightblue', mec='black') 
# Platinum
plt.plot_date(x=df['Date'], y=df['Platinum'], linestyle='--', c='orange') 
```
